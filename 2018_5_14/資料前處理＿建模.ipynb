```python
import numpy as np 
from scipy.io import loadmat
import keras 
import pandas as pd

#用 pandas 讀取 csv
df1 = pd.read_csv("train.csv")

#train.csv 資料長這樣
df1.head()
```

```python
df1.shape
```

```python
# 讀取 train.csv 中欄位 Activity 的資料 （這是要 y_train 要作為分類的結果）
y_train = df1["Activity"]
print(type(y_train))

#把y_train 變成 numpy array 也可以用 .as_matrix  
y_train= y_train.values  
print(type(y_train))

#把 y_train 裡的中文變成數字作為其代號
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y_train=le.fit_transform(y_train)
print(y_train) 
    

```

```python
#讀取 train.csv 中，除了欄位 Activity以外的資料
x_train = df1.drop("Activity",axis=1)
#
x_train = x_train.values
```

# 處理test.csv 分成 x_test , y_test

```python
df2 = pd.read_csv("test.csv")
df2.head()
```

```python
x_test = df2.drop("Activity",axis =1) #原始預設是drop row 所以要寫axis = 1 to drop column 
y_test = df2["Activity"]

x_test= x_test.values
y_test= y_test.values

#把 y_test 裡的中文變成數字作為其代號
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y_test = le.fit_transform(y_test)
print(y_test) 

```

```python
#印出形狀 
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

```python

# 開始建模

import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten

num_classes = 6 #

# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
model = Sequential()
model.add(Dense(100, input_shape = (562, ), activation="relu")) 
model.add(Dense(6, activation='softmax'))

model.summary()
```

```python
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer="adam",
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train,
          batch_size=32,
          epochs=3,
          verbose=1,
          validation_data=(x_test, y_test))
```
